# conclusion 

```python
#plotting profit v/s sales statewise
profit_to_sales = pd.DataFrame(data.groupby('State').sum()[['Sales','Profit']])
profit_to_sales['Profit_to_sales'] = profit_to_sales['Profit']/profit_to_sales['Sales']
profit_to_sales.sort_values(by='Profit_to_sales',inplace = True, ascending = False)
print("States with the highest 10 Ratios :\n")
print(profit_to_sales[:10])
print('*'*70)
print("States with the Lowest 10 Ratios:\n")
print(profit_to_sales[-10:])
```

*states district of columbia ,delware and minesota generates profit to sales ration but the buisness should be expanded more in states like florida,oregon and ohio to maximise the profit*

```python
#plotting profit v/s sales region wise
profit_to_sales = pd.DataFrame(data.groupby('Region').sum()[['Sales','Profit']])
profit_to_sales['Profit_to_sales'] = profit_to_sales['Profit']/profit_to_sales['Sales']
profit_to_sales.sort_values(by='Profit_to_sales',inplace = True, ascending = False)
print("Region wise profit v/s sales analysis:\n")
print(profit_to_sales[:10])
```

*the southern and central regions should be focused to generate more profit*

```python
 #plotting profit v/s sales sub-category wise
profit_to_sales = pd.DataFrame(data.groupby('Sub-Category').sum()[['Sales','Profit']])
profit_to_sales['Profit_to_sales'] = profit_to_sales['Profit']/profit_to_sales['Sales']
profit_to_sales.sort_values(by='Profit_to_sales',inplace = True, ascending = False)
print("items with the highest 10 Ratios :\n")
print(profit_to_sales[:10])
print('*'*70)
print("items with the Lowest 10 Ratios:\n")
print(profit_to_sales[-10:])
```

*tabels,bookcases and supplies bring huge sales but they are not profitable*

```python
#plotting profit v/s sales region wise
profit_to_sales = pd.DataFrame(data.groupby('Category').sum()[['Sales','Profit']])
profit_to_sales['Profit_to_sales'] = profit_to_sales['Profit']/profit_to_sales['Sales']
profit_to_sales.sort_values(by='Profit_to_sales',inplace = True, ascending = False)
print("category wise profit v/s sales analysis:\n")
print(profit_to_sales[:10])
```

*major buisness is generated from the office supplies but the technology and Furniture items should be focused more to maximise the buisness*
